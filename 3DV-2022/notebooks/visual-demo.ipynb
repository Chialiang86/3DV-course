```python
import matplotlib.pyplot as plt
import numpy as np


voxels = np.load('../data/chair_img_pc_voxel_mesh/1a6f615e8b1b5ae4dbbc9440457e303e/voxel.npy')
print(f'voxels : {voxels.shape}')

points = np.load('../data/chair_img_pc_voxel_mesh/1a6f615e8b1b5ae4dbbc9440457e303e/point_cloud.npy')
points = points - np.expand_dims(np.mean(points, axis = 0), 0) # center
dist = np.max(np.sqrt(np.sum(points ** 2, axis = 1)),0)
points = points / dist #scale
print(points.shape)
fig = plt.figure()
ax = fig.add_subplot(projection='3d')

ax.scatter(points[...,0], points[...,2], points[...,1], c='r', marker='.')

ax.set_xlabel('X Label')
ax.set_ylabel('Y Label')
ax.set_zlabel('Z Label')

plt.show()
```
